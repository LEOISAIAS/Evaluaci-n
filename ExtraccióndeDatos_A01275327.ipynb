```python
#Importamos las librerias pandas, numpy y matplotlib respectivamente
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#**EUA**

```python
EUA = pd.read_csv("EUAlimpio.csv")
EUA
```

## FILTRO 1

```python
filtro1 = EUA.loc[EUA['host_acceptance_rate'] > 50]
filtro1
```

# FILTRO 2

```python
filtro2 = EUA.loc[EUA['host_is_superhost'] == 't']
filtro2
```

# FILTRO 3

```python
filtro3 = EUA.loc[EUA['host_identity_verified'] == 'f']
filtro3
```

# FILTRO 4

```python
filtro4= EUA.loc[EUA['property_type'].isin(['Private room', 'Hotel room'])]
filtro4
```

# FILTRO 5

```python
filtro5 = EUA.loc[EUA['bathrooms'] > 1]
filtro5
```

# FILTRO 6

```python
filtro6 = EUA.loc[(EUA['price'] > 10000) & (EUA['property_type'] == 'Entire home')]
filtro6
```

# FILTRO 7

```python
filtro7= EUA.loc[EUA['review_scores_cleanliness'] > 4.5]
filtro7
```

# FILTRO 8

```python
filtro8 = EUA.loc[EUA['review_scores_value'] > 4.9]
filtro8
```

# FILTRO 9


```python
filtro9 = EUA.loc[EUA['availability_365'] < 100]
filtro9
```

# FILTRO 10

```python
filtro10 = EUA.loc[EUA['host_response_time'] == 'within an hour']
filtro10
```

```python
#Convertir DataFrame a CSV
filtro1.to_csv("Filtro1_EUA.csv")
filtro2.to_csv("Filtro2_EUA.csv")
filtro3.to_csv("Filtro3_EUA.csv")
filtro4.to_csv("Filtro4_EUA.csv")
filtro5.to_csv("Filtro5_EUA.csv")
filtro6.to_csv("Filtro6_EUA.csv")
filtro7.to_csv("Filtro7_EUA.csv")
filtro8.to_csv("Filtro8_EUA.csv")
filtro9.to_csv("Filtro9_EUA.csv")
filtro10.to_csv("Filtro10_EUA.csv")


#descargar archivo filtrado en csv
from google.colab import files 
  
  
files.download("Filtro1_EUA.csv")
files.download("Filtro2_EUA.csv")
files.download("Filtro3_EUA.csv")
files.download("Filtro4_EUA.csv")
files.download("Filtro5_EUA.csv")
files.download("Filtro6_EUA.csv")
files.download("Filtro7_EUA.csv")
files.download("Filtro8_EUA.csv")
files.download("Filtro9_EUA.csv")
files.download("Filtro10_EUA.csv")
```
